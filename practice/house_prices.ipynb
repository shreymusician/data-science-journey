### Level - 2 Questions

#### 5) Which city has the highest average house price, and what is that average price?

```python
df.head()
```

```python
# Using Method - Chaining Method
(
    df
    .groupby('city')['price']
    .mean()
    .round(2)
    .to_frame('Average House Price')
    .sort_values(by='Average House Price', ascending=False)
    .iloc[:1]
)
```

```python
# Another method
avg_df = df.groupby('city')['price'].mean().round(2)
city = avg_df.idxmax()
value = avg_df.max()

print(f'City : {city}, Avg Price : {value}')
```

#### 6) For each different type of property (Residential, Condo, etc.), find both the minimum price and the maximum price.


```python
df.groupby('type')['price'].agg(['min', 'max'])
```

### Level - 3 Questions

#### 7) Find all the properties that are Condos AND have a price greater than $300,000. Return a DataFrame showing only their street, city, and price.

```python
(
    df[ (df['type'] == 'Condo') & (df['price'] > 300000) ]
    [['street', 'city', 'price']]
    .sort_values('price')
)
```

### Level - 4 Questions

#### 8) Create a new column in the dataframe called price_per_sqft. Then, find the top 5 streets with the highest price_per_sqft.

```python
# Using Broad-Casting
df['price_per_sqft'] = (df['price'] / df['sq__ft']).round(2)
df = df[df['sq__ft'] > 0]
df= df.sort_values(by='price_per_sqft', ascending=False)
df.iloc[:5]
```

### Level - 5 Questions

#### Question 9

**For each city, we want to find out two things:** <br>
1) What is the total number of transactions (count) that took place? <br>
2) What is the median price_per_sqft for that city? <br>

**Bonus: Sort the final result so that the city with the highest median price_per_sqft is at the very top.**

```python
res = df.groupby('city')['price_per_sqft'].agg(['count', 'median']).round(2)
res.columns = ['Total Transactions', 'Median Price Per Sqft']
res = res.sort_values(by='Median Price Per Sqft', ascending=False) 
res.head()
```

### Level - 6 Questions

#### 10) Convert the sale_date column into a proper pandas datetime data type. Then, print out df['sale_date'].dtype to confirm it changed.

```python
ndf = pd.read_csv('house_prices.csv')
ndf.head()
```

```python
ndf['sale_date'].dtype
```

```python
ndf['sale_date'] = pd.to_datetime(ndf['sale_date'])
ndf.head()
```

#### Question 11: Create a new column called day_of_week that contains the actual name of the day (e.g., Monday, 
#### Tuesday) for each transaction. Then, count how many houses were sold on each day of the week.

```python
ndf['day_of_week'] = ndf['sale_date'].dt.day_name()
ndf.head()
```

```python
ndf.groupby('day_of_week').size().to_frame('No. of Transactions')
```

#### 12) Which specific date (sale_date) generated the highest total revenue (the sum of all house prices combined), and what was that total amount?

```python
(
    ndf
        .groupby('sale_date')['price']
        .agg('sum')
        .to_frame('Total Revenue')
        .sort_values(by='Total Revenue', ascending=False)
        .iloc[:1]
)
```

#### Question 13: Group the data by the day of the month (e.g., 1st, 15th, 21st) and calculate the average (mean) price for each day.
*Bonus: Sort the index so it reads chronologically from the earliest day of the month to the latest.*

```python
(
    ndf
        .groupby(ndf['sale_date'].dt.day)
        ['price']
        .agg('mean')
        .round(2)
        .to_frame('Mean Price')
        .sort_index()
)
```

#### Question 14: Group the data by the exact sale_date and find the maximum property price sold on each day. Sort it chronologically by the date.

*Bonus Thought: Look closely at your final result table. Is there one specific day where the maximum price looks wildly out of place compared to the others?*

```python
ndf.head()
```

```python
(
    ndf
        .groupby(ndf['sale_date'])
        ['price']
        .agg('max')
        .round(2)
        .to_frame('Max Price')
        .sort_index()
)

# Unusual date -> 2008-05-15
```

#### Question 16: Group the data by both day_of_week and type. Find the total number of transactions and the average price for each combination.

*Bonus: Round the average price to 2 decimal places.*

```python
(
    ndf.groupby(['day_of_week', 'type'])
    ['price']
    .agg(['count', 'mean'])
    .round(2)
)
```

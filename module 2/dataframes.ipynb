### DataFrame
It is a 2-D object containing index and multiple columns of content, with each column having a label.<br>
**DataFrame = Two-Axis Labelled Array**<br>
Part of Pandas Library

```python
import pandas as pd

# 3 school records
r1 = pd.Series({'Name' : 'Alice', 'Class' : 'Physics', 'Score' : 85})

r2 = pd.Series({'Name' : 'Jack', 'Class' : 'Chemistry', 'Score' : 82})

r3 = pd.Series({'Name' : 'Helen', 'Class' : 'Biology', 'Score' : 90})
```

```python
# DataFrame object is indexed
# DataFrame represents group of series where each series represents a row of data
# Just like series, we can pass an array and index values as second arguments

df = pd.DataFrame([r1,r2,r3], index=['s1','s2','s3'])
df
```

```python
# Alternative is to use Dictionary
fd = [
{'name':'Shreyas','alias':'Turtle'},
{'name': 'Deekshith','alias':'Kardi'},{'name':'Zain','alias':'Xain Grylls'},{'name':'Navya','alias':'Nivea'},{'name':'Sumaiya','alias':'Suman'},{'name':'Tejaswini','alias':'Turtle'}]

f = pd.DataFrame(fd)
f
```

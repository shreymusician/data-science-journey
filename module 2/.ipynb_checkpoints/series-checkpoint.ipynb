### Series Datatype 
Series object stores Single Dimensional Indexed Data<br>
It is often similar to both List and Dictionary

```python
import pandas as pd

# We can create series by passing a list
# Default starting index -> 0 and default name -> None
stud = ['Shreyas', 'Deekshith', 'Zain']
pd.Series(stud)
```

```python
# Series of numbers
pd.Series([2,6,8,16])
```

```python
# Pandas handle empty data -> NaN values
stud = ['Shreyas', 'Zain', None]
pd.Series(stud)
```

```python
# Same for integer values
pd.Series([1,2,None]) # Converted into float when added None
```

```python
# NaN is not equivalent to None
import numpy as np
np.nan == None
```

```python
# Can't do comparision on NaN
print(np.nan == np.nan)

# You need to use special function to check if it is NaN value
print(np.isnan(np.nan))
```

### Methods of creating Series

```python
# Can be done using List
# Better to use dictionary for more detailed data
interests = {'Shreyas':'Music', 'Zain':'Anime','Deekshith':'Running'}
i = pd.Series(interests)
print(i)

# Keys becomes Index Values
print('\n',i.index)
```

```python
# List of tuples
stud = [('Shreyas','Music'), ('Deekshith', 'Running'), ('Zain', 'Books')]
pd.Series(stud)
```

```python
# Custom Index
pd.Series(['Music', 'Running', 'Books'], index=['Shreyas', 'Deekshith', 'Zain'])
```

```python
# Custom Index always overrides Dictionary Index
# Pandas provides NaN values for any custom index not in dictionary
con = {'Male':'Wizard', 'Female':'Witch'}
s = pd.Series(con)
print(s)

# Using custom index values
t = pd.Series(con, index=['Male', 'Female', 'Child'])
print('\n',t)
```

### Querying Series

```python
# Learn how to query and merge series
# Pandas Series can be queried either by index position or index label
# To Query by index position starting from 0 we use iloc
# To Query by index label we use loc

import pandas as pd
teach = {'Maths' : 'Charan', 'Chemistry' : 'Pallavi', 'Kannada' : 'Geetha'}
t = pd.Series(teach)
print(t.iloc[0])
print(t.loc['Chemistry'])

# iloc and loc are not methods, they are attributes
```

### Working with data in series

```python
# Iterate over all items and apply the operation
fd = {'Shreyas':'Turtle', 'Deekshith':'Kardi', 'Zain':'Xain Grylls', 'Navya':'Nivea', 'Sumaiya':'Suman', 'Tejaswini':'Teju'}
f = pd.Series(fd)
print(f)

# Vectorization works with most of functions in Numpy
# It means running many instructions at once using High Performance H/w like GPU

# Series Datatype is Iterable hence can be passed into numpy
c = [9.1,9,8,8.5,8.8,8.4]

cf = pd.Series(c, index=fd.values())
print('\n',cf)

avg = np.mean(cf)
print(f"\nAverage CGPA : {avg:.2f}")
print("Highest CGPA : ", np.max(cf))
print("Lowest CGPA : ", np.min(cf))
```

```python
# Using random numbers
num = pd.Series(np.random.randint(0,1000,10000))
print(num.head()) # first five
print('\n',num.tail()) # last five
print('\nTotal Length : ', len(num))
```

```python
#Ipython has Magic Functions begin with % sign, use Tab key
# Cellular Magic Function
# Timeit - Runs code few times to find how long it takes
# It must begin with a new cell
```

```python
%%timeit -n 100
total = 0
for each in num:
    total += each
total/len(num)
```

```python
%%timeit -n 1000
total = np.sum(num)
total / len(num)
```

### BroadCasting

```python
# Operation to every value in series
print(num.head())
num+=2
print('\n',num.head())

# Broadcasting is a shortcut way of looping
```

```python
# loc helps not only to modify data in place but also add new data
s = pd.Series([1,2,3])
print(s)

# Adding a new label using loc
s.loc['New'] = 4
print('\n',s)

# Hence the labels and values for Series can have mixed datatypes
s.loc['Five'] = 'Five'
print('\n',s)
```

```python
# Interesting feature -> Same data label with different values
print(f)

# Create a new Series for a new Friend
k = pd.Series(['Kona','Kempi'],index=['Karthik','Karthik'])
print('\n',k)

# Merge two series using pd.concat([s1,s2]) creates a new series
nf = pd.concat([f,k])
print('\n',nf)

# When we Query, we get complete values
print('\n',nf.loc['Karthik'])
```

### A Fun Hashing Application :)

```python
n = int(input("Enter number of values : "))
vl = np.random.randint(1,300,n)

print("\nValue List : ", vl)

hash_table = {}

for v in vl:
    key = v%n
    
    while key in hash_table:
        key = (key + 1) % n
        
    hash_table[key] = v

s = pd.Series(hash_table)
s = s.sort_index()
print('\n', s)
```

```python
# We can give custom name to Series Data
fam = pd.Series(['shreyas', 'siri', 'sangi', 'satya'], index=['son', 'daughter', 'mother', 'father'], name='Family')
print(fam)

# We can easily inspect the range of index values
print('\n',fam.index)

# Add a NaN value
fam.loc['grandpa'] = None
print('\n', fam)

# Pandas ignore NaN values
print('\nNo. of entries :',fam.count())

# Creating a boolean array to determine null values
arr = [pd.isnull(ele) for ele in fam] # can use pd.isna()
print('\n',arr)
```

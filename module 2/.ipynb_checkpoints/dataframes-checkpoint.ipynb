### DataFrame
It is a 2-D object containing index and multiple columns of content, with each column having a label.<br>
**DataFrame = Two-Axis Labelled Array**<br>
Part of Pandas Library<br>
Replicates Database

```python
import pandas as pd

# 3 school records
r1 = pd.Series({'Name' : 'Alice', 'Class' : 'Physics', 'Score' : 85})

r2 = pd.Series({'Name' : 'Jack', 'Class' : 'Chemistry', 'Score' : 82})

r3 = pd.Series({'Name' : 'Helen', 'Class' : 'Biology', 'Score' : 90})
```

```python
# DataFrame object is indexed
# DataFrame represents group of series where each series represents a row of data
# Just like series, we can pass an array and index values as second arguments

df = pd.DataFrame([r1,r2,r3], index=['s1','s2','s3'])
df
# df.head()
```

```python
# Alternative is to use Dictionary
import pandas as pd

fd = [
{'name':'Shreyas','alias':'Turtle'},
{'name': 'Deekshith','alias':'Kardi'},{'name':'Zain','alias':'Xain Grylls'},{'name':'Navya','alias':'Nivea'},{'name':'Sumaiya','alias':'Suman'},{'name':'Tejaswini','alias':'Turtle'}]

f = pd.DataFrame(fd)
print(f)

# We can use iloc[] and loc[] for indexing just like Series
print('\n',f.iloc[-1])
```

```python
import pandas as pd

data = {
'index' : ['shreyas','siri', 'sangi', 'satya'],
'data' : ['aiml', 'pcmb', 'cbz', 'eee']
}

df = pd.DataFrame(data, index=['child','child','mom','dad'])

print(df)

# Index using loc[]
print('\n', df.loc['mom'])
```

```python
# Adv of df is we can select data based on multiple axes

print(df.loc['child','index'])

# Hence, repetition is allowed in df both in row and col labels
```

```python
# Transposing df (pivot)
df.T.loc['index'] # Permanent operation
```

```python
print(df)

# Selecting specific rows and cols only
print('\n',df.loc['child']['index']) # Avoid
```

```python
# Slicing can be applied to loc[]
print(df.loc[:])

# Syntax -> .loc[row,col]
print('\n', df.loc[:,['index']])
```

#### Selection Operation
df[df["Branch"] = "AIML"]

#### Projection Operation
df[["Name","Branch"]]

```python
# Dropping data
# Using drop function takes row as argument, and returns a changed df

print(df)

print('\n',df.drop('child',inplace=False))
```

```python
# drop has two optional param
# inplace -> Changes applied inplace
# axis -> default row (0), col(1)

# Returns a view

# Initial df
print(df)

# Dropping child rows inplace
print('\n',df.drop('child')) 

# Dropping columns
print('\n', df.drop('data',axis=1))
```

```python
# Another way of deleting is using del
print(df)

del df['data'] # deletes col
print('\n',df)
```

```python
# Adding new col is easy
df['data'] = ['aiml','pcmb','cbz','eee']
print(df)
```

### Querying a DataFrame

```python
# Using boolean masking we can query df
df = pd.read_csv('datasets/Admission_Predict.csv', index_col=0)

df.columns = [x.lower().strip() for x in df.columns]

df.head()
```

```python
# Having easier names
df.columns = ['gscore','tscore','rate','sop','lor','gpa','res','chance']

df.head()
```

```python
# Students having chance > 0.7
df[df['chance'] > 0.7].sort_values(by='chance').head()
```

```python
# Using Boolean Masking
# Student having chance > 0.7
mask = df['chance'] > 0.7
mask.head()
```

```python
# Filtering out values by layering the mask over the original df
df.where(mask).head() # True Masking
```

```python
# Use dropna() to remove NaN fields
df.where(mask).dropna().head()
```

```python
# Selecting operation
# Select all rows who have not done research and order by gpa increasing

df[df['res'] == 0].sort_values(by='gpa').head()
```

```python
# Projecting operation
df[df['res'] == 0][['res','gpa','chance']].sort_values(by='gpa').head()
```

#### Boolean Masking Operation (BitMasking)

mask1 & mask2<br>
mask1 | mask2<br>
~ mask

```python
df.head()
```

```python
# Combining two conditions
# Suppose we want to print out students who have done research and have gpa > 9

stu = df[ (df['res'] == 1) & (df['gpa'] > 9) ][['res','gpa','chance']].sort_values(by='gpa')

stu.head()
```

```python
# Using built-in operators to avoid parenthesis

stu = df[df['res'].eq(1) & df['gpa'].gt(9)][['res','gpa','chance']].sort_values(by='gpa')
stu.head()
```

```python
# I want to find out chance of top students having gpa >= 8 and <= 10

top = df[df['gpa'].ge(8) & df['gpa'].le(10)][['gpa','chance']].sort_values(by='gpa')
top.head()
```

### Indexing DataFrames

```python
# Using set_index() -> Destructive - Destroys existing index

df = pd.read_csv('datasets/Admission_Predict.csv', index_col = 0)
df.head()
```

```python
# Suppose we want to change index form Serial No. to Chance of Admission

# Copy indexed data into its own column
df['Serial No.'] = df.index

df.columns = [x.lower().strip() for x in df.columns]

# Set index into another column
df = df.set_index('chance of admit')

df.head()
```

```python
# Get rid of index by using reset_index(), creates default indexes

df = df.reset_index()
df.head()
```

```python
# Pandas supports Multi-level indexing
# Similar to composite keys in RDBMS
# Using set_index() and input list of columns

df = pd.read_csv('datasets/census.csv')
df.head()
```

```python
# Cleaning the column names
df.columns = [x.lower().strip() for x in df.columns]

df.head()
```

```python
# Shape of the table
df.shape
```

```python
# Description of the table
df.describe().loc[['min','max','mean']].round(2)
```

```python
# To find list of unique values in sumlev
df['sumlev'].unique()
```

```python
# Creating a dict of unique values
ud = {}

for ele in df.columns:
    ud[ele] = [len(df[ele].unique()), df[ele].unique().min(), df[ele].unique().max()]

udf = pd.DataFrame(ud, index=['Unique Elements', 'Min', 'Max'])
udf.head()
```

```python
# Selection operation over sumlev == 50
df[df['sumlev'].eq(50)].head()
```

```python
columns_to_keep = ['STNAME','CTYNAME','BIRTHS2010','BIRTHS2011','BIRTHS2012','BIRTHS2013',
                   'BIRTHS2014','BIRTHS2015','POPESTIMATE2010','POPESTIMATE2011',
                   'POPESTIMATE2012','POPESTIMATE2013','POPESTIMATE2014','POPESTIMATE2015']

columns_to_keep = [x.lower() for x in columns_to_keep]

df = df[columns_to_keep]
df.head()
```

```python
# As we discussed before, we can have composite keys in df just like in RDBMS

df = df.set_index(['stname','ctyname'])
df.head()
```

```python
# How to query df having composite index
df = df.sort_index() # sort the values in ascending order such that searching performance is increased

df.loc[('Alabama','Alabama')]
```

### Example


```python
df = pd.read_csv('datasets/presidents.csv')
df.set_index('#', inplace=True)
df.head()
```

```python
# Separating President's name into First Name and Last Name using Regex

df['First Name'] = df['President']
df['First Name'] = df['First Name'].replace("[ ].*", "", regex=True)
df.head()
```

```python
# Better Method to separate President's Name

del df['First Name']

# Create function to split name
def splitname(row):
    row['First Name'] = row['President'].split(' ')[0]
    row['Last Name'] = row['President'].split(' ')[-1]
    return row

df = df.apply(splitname, axis='columns')
df.head()
```

```python
# extract()
del df['First Name']
del df['Last Name']

```

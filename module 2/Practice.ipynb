```python
import pandas as pd
df = pd.read_csv('netflix_titles.csv')
df.head()
```

```python
# Remove unwanted columns
df.drop(columns=['listed_in', 'description', 'cast'], inplace=True)

# Set index as show_id
df.set_index('show_id', inplace=True)
df.head()
```

```python
# Find all Meta-Data of the Dataset

print(f'Rows : {df.shape[0]}, Columns : {df.shape[1]}')
print(f'Total Data : {df.size}')
print('\n\nInfo : \n')
df.info()
```

```python
# All rows and columns
cols = df.columns.tolist()
idx = df.index.tolist()

print(f'\nColumn data : {cols}')
```

```python
df.head()
```

```python
# Converting to Date-Time Datatype
df['date_added'] = pd.to_datetime(df['date_added'], format = 'mixed')
df.head()
```

```python
df = df[['title', 'type', 'country', 'release_year', 'director', 'rating', 'duration', 'date_added']]
df.head()
```

```python
# Find unique values in each column
udf = df.nunique().to_frame('Number of Unique elements').sort_values(by = 'Number of Unique elements')
udf
```

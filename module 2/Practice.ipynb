```python
# Which decade produced max content
# Let's create a new table
dec = df['release_year'].copy().to_frame()

# Function to Calculate Decade
def to_decade(year):
    syear = str(year)
    return(int(syear[:3] + '0'))

dec['decade'] = dec['release_year'].apply(to_decade)
dec.tail()
```

```python
# Find answer - 2010s produced maximum content
dec[['decade']].groupby('decade').value_counts().sort_values().iloc[-1:]
```

```python
# Find the country whose average release year is the highest

# Create a new df
cdf = df[['country', 'release_year']].copy()
cdf.dropna(inplace=True)
cdf.head()
```

```python
# Split Country
cdf['country'] = cdf['country'].str.split(',')
cdf.head()
```

```python
# Explode Country
cdf = cdf.explode('country')
cdf['country'] = cdf['country'].str.strip()
cdf.head(15)
```

```python
# Using groupby to find answers
netyears = len(cdf)

res = cdf.groupby('country').size().to_frame('Average years')
res['Average years'] = (res['Average years'] / netyears).round(2)
res = res.sort_values(by='Average years', ascending=False)
res

# United States is the country with highest average release years
```

```python
# Among titles added in 2021, which rating appears most often?

# Create a separate df
rdf = df['rating'].to_frame()
rdf['year_added'] = df['date_added'].dt.year
rdf.dropna(inplace=True)
rdf['year_added'] = rdf['year_added'].astype(int)
rdf.tail(10)
```

```python
# Keep only values where year_added == 2021
rdf = rdf[ rdf['year_added'] == 2021 ]
rdf.head(20)
#len(rdf)
```

```python
# Use Group-By to find answers
res = rdf.groupby('rating').size().to_frame('Freq')
res = res.sort_values(by='Freq', ascending=False)
res

# Among titles added in 2021, "TV-MA" rating appears most often
```

```python
# Find the longest movie

# Create a new df
mdf = df[df['type'] == 'Movie']
mdf.dropna(inplace=True) # Remove NaN values
mdf.head()
```

```python
# Manipulate duration column to make it int type

# Create a function
def change(dur):
    res = dur.split(' ')[0]
    return res

# Apply the function
mdf['duration'] = mdf['duration'].apply(change)

# Now change the datatype of duration to int
mdf['duration'] = mdf['duration'].astype(int)

# Sort in descending order
mdf = mdf.sort_values(by='duration', ascending=False)
mdf.head()
```

```python
# Find the answer
mdf[mdf['duration'] == mdf['duration'].max()]
```

```python
'''
Create a DataFrame that contains only:

Country
Number of Movies
Number of TV Shows
Total Titles

sorted by Total Titles
'''

# Create a new df
sol = df[['type', 'country']]
sol = sol.dropna()
sol.head()
```

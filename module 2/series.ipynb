### Series Datatype 
Series object stores Single Dimensional Indexed Data<br>
It is often similar to both List and Dictionary

```python
import pandas as pd

# We can create series by passing a list
# Default starting index -> 0 and default name -> None
stud = ['Shreyas', 'Deekshith', 'Zain']
pd.Series(stud)
```

```python
# Series of numbers
pd.Series([2,6,8,16])
```

```python
# Pandas handle empty data -> NaN values
stud = ['Shreyas', 'Zain', None]
pd.Series(stud)
```

```python
# Same for integer values
pd.Series([1,2,None]) # Converted into float when added None
```

```python
# NaN is not equivalent to None
import numpy as np
np.nan == None
```

```python
# Can't do comparision on NaN
print(np.nan == np.nan)

# You need to use special function to check if it is NaN value
print(np.isnan(np.nan))
```

### Methods of creating Series

```python
# Can be done using List
# Better to use dictionary for more detailed data
interests = {'Shreyas':'Music', 'Zain':'Anime','Deekshith':'Running'}
i = pd.Series(interests)
print(i)

# Keys becomes Index Values
print('\n',i.index)
```

```python
# List of tuples
stud = [('Shreyas','Music'), ('Deekshith', 'Running'), ('Zain', 'Books')]
pd.Series(stud)
```

```python
# Custom Index
pd.Series(['Music', 'Running', 'Books'], index=['Shreyas', 'Deekshith', 'Zain'])
```

```python
# Custom Index always overrides Dictionary Index
# Pandas provides NaN values for any custom index not in dictionary
con = {'Male':'Wizard', 'Female':'Witch'}
s = pd.Series(con)
print(s)

# Using custom index values
t = pd.Series(con, index=['Male', 'Female', 'Child'])
print('\n',t)
```

### Querying Series

```python
# Learn how to query and merge series
# Pandas Series can be queried either by index position or index label
# To Query by index position starting from 0 we use iloc
# To Query by index label we use loc

import pandas as pd
teach = {'Maths' : 'Charan', 'Chemistry' : 'Pallavi', 'Kannada' : 'Geetha'}
t = pd.Series(teach)
print(t.iloc[0])
print(t.loc['Chemistry'])

# iloc and loc are not methods, they are attributes
```

### A Fun Hashing Application :)

```python
n = int(input("Enter number of values : "))
vl = np.random.randint(1,300,n)

print("\nValue List : ", vl)

hash_table = {}

for v in vl:
    key = v%n
    
    while key in hash_table:
        key = (key + 1) % n
        
    hash_table[key] = v

s = pd.Series(hash_table)
s = s.sort_index()
print('\n', s)
```

```python
# We can give custom name to Series Data
fam = pd.Series(['shreyas', 'siri', 'sangi', 'satya'], index=['son', 'daughter', 'mother', 'father'], name='Family')
print(fam)

# We can easily inspect the range of index values
print('\n',fam.index)

# Add a NaN value
fam.loc['grandpa'] = None
print('\n', fam)

# Pandas ignore NaN values
print('\nNo. of entries :',fam.count())

# Creating a boolean array to determine null values
arr = [pd.isnull(ele) for ele in fam] # can use pd.isna()
print('\n',arr)
```

```python
# Masking technique used in Numpy
a = np.random.randint(1,25,12)
print('Initial elements : ',a)
print('Mean : ',a.mean())

mask = a < a.mean()
print('Mask array : ',mask)
print('\nElements lesser than mean : ',a[mask])

# We can create a mask array based on specific condition, and later filter the elements out in the original array
# Short-hand method to apply masking
print('Short-hand method : ',a[a < a.mean()])
```

```python
# It is better to obtain a new series by performing an operation than updating series in place
# If you need to specify multiple values for same index names, it is better to pass two lists, than dict

# Checking Membership in Series Data
s = pd.Series([13,15,17,19])
print(s)

# Checking if 17 is an element of Series
print('\n',17 in s) # This returns false even if 17 is an element
# Note : 'in' operator works for index labels not for values
print(2 in s) # returns true for index

# Proper way to Query membership of a value in Series
print('\n', 17 in set(s)) # Use set datatype
print(17 in s.values) # Use values attribute
```

```python
print(s,'\n')

# To iterate over both index labels and values use items()
for data in s.items():
    print(data) # Returns data as tuples
print('\n')

# If we want individual labels and values
for i,v in s.items():
    print(f"Index : {i}, Value : {v}")
```

```python
# Updating Values in Series
print(s)

# Can be done using index labels 
s[3] = 20
print('\n',s)

# Indexing can also be used to add new values to the Series
s[4] = 100
print('\n',s)
```

```python
# Using indexing can either update or append new data, one must be careful !!!
t = pd.Series([3,4,5,6,7,8], index=[1,2,3,4,4,4]) # For same index, many values
print(t)

t[4] = 500 # Updates all instances
print('\n',t)
```

Hence, for Updating Existing Data or Adding New Data in Series, it is much safer to use **loc[] and iloc[]** methods 

```python
t = pd.Series([3,4,5,6,7,8], index=[1,2,3,4,4,4]) # For same index, many values
print(t)

# Using iloc[] we can change only a specific entry, (7 to 500)
t.iloc[4] = 500
print('\n',t)
```

### Deletion in Series
Deletion is generally not used in Pandas, rather new Data Structure is obtained my Modifying or Altering existing data using **Masks**.
<br>We can use ***del*** keyword to delete an entry

```python
# Delete the entry 
s = pd.Series([1,2,3,4])
print(s)

del s[1]
print('\n',s)
```

### Using iloc[] and loc[]
iloc[] -> based on indexing by position <br>
loc[] -> based on indexing by label

```python
# We can Slice a series using iloc[] to form a new series
import pandas as pd

a = pd.Series([5,10,15,20,25])
print(a)

b = a.iloc[:3]
print('\n',b)

# If we want to choose specific index values only, we can pass in a list of index values
c = a.iloc[[0,2,4]]
print('\n',c)
```

```python
a = pd.Series([5,10,15,20,25], index=['first','second','third','fourth','fifth'])
print(a)

# Slicing using loc[]
b = a.loc[:'fourth'] # inclusion on both sides
print('\n',b)

# We can pass a list having specific index values
c = a.loc[['second','fourth']]
print('\n',c)
```

***.at[] and .iat[]*** are other index accessors that are similar to ***.loc[] and .iloc***<br>

Only difference is that they return Numpy arrays instead of Series.<br>

Also they do not support **Slicing** and do not support passing **list of Index values**.<br>

They work only for a **Single Index Value**

```python
a = pd.Series([5,10,15,20,25])
print(a)

# Using .at[]
b = a.at[4]
print('\n',b)
```

```python
a = pd.Series([5,10,15,20,25], index=['first','second','third','fourth','fifth'])
print(a)

# Using .at[]
b = a.at['first']
print('\n',b)
```

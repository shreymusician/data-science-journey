```python
# Python has null param capabilities

def add(x,y,z=0):
    print(x+y+z)

add(1,2)
add(1,2,3)
```

```python
# Assign values to functions

def add(x,y,z=0,flag=False):
    if(flag):
        print("Flag is True")
    
    else:
        print("Flag is False")
    
    print(x+y+z)

add(1,2)
add(1,2,3,flag=True)
```

```python
# Another important datastructure is Dictionary, which is similar to Map, We store values, and to access it we require keys

d = { 'AIML' : 'Sunitha', 'DS' : 'Adarsh'}

d['AIML']

d['MECH'] = None
print(d)

d['MECH'] = 'Satya'
print(d)
```

```python
# Iterating Dict

# Key-Value Pairs
for key in d:
    print(f" key : {key}, value : {d[key]}")
print('\n')

# Key-Values Pairs using items()
for k,v in d.items():
    print(f" key : {k}, value : {v}")
print('\n')

# Only Values
for val in d.values():
    print(val)
print('\n')
```

```python
# CSV File Manipulation

import csv
%precision 2

with open('mpg.csv') as csvfile:
    mpg = list(csv.DictReader(csvfile))
    
#mpg[:3]
print("Total Cars :", len(mpg))
```

```python
# Use Set datatype anywhere you want unique values to be chosen

cyldata = set(d['cyl'] for d in mpg)
print(cyldata)

cylcount = {}
for val in cyldata:
    cylcount[val] = 0

print(cylcount)

for k,v in cylcount.items():
       for d in mpg:
            if k == d['cyl']: 
                cylcount[k] += 1

print(cylcount)
```

```python
vclass = set(d['class'] for d in mpg)
print(vclass)
```

```python
classcount = {}

for item in vclass:
    classcount[item] = 0

print(classcount)

for k in classcount:
    for d in mpg:
        if d['class'] == k:
            classcount[k] += 1

print(classcount)
```

### Regex for Data Cleaning

```python
# Regex - Regular Expression, used to clean textual data. We can write our own regex expression
import re

# To find pattern string at starting 
main = "This is a Good Day :)"
if re.match("This",main):
    print("Yes")

# To find anywhere in the main string
if re.search("Good",main):
    print("Yes")
```

```python
# Just like checking using Conditionals, we can Segment a String.
# Tokenizing -> Splitting a string into substrings based on patterns.
# Tokenizing is used heavily in NLP

# Using split()
main = "Shreyas is a Coder. Shreyas is also a Musician. Shreyas has a great life."
print(re.split('Shreyas',main))

# Using findall()
print(re.findall('Shreyas', main))
```

```python
# ^ -> Starting | $ -> Ending
re.search("^Shreyas", main)
re.search("a great life.$", main)
```

### Patterns and Character Classes

```python
grades = "AABBBAABBBBBBAAABCCCCCAACABDCC"

# Single Grade
print(re.findall("B",grades))

# Multiple Grades using Set Operator (Square Brackets)
print(re.findall("[BC]",grades))

# Combination of Grades
print(re.findall("[B][A-C]",grades))
print(re.findall("[A][B][C]",grades))

# Using Pipeline operator
print(re.findall("AB|AC",grades))

# Except some character
print(re.findall("[^B]",grades))

# Special Query
print(re.findall("^[^A]",grades)) # returns all characters as a list that starts with not 'A'
```

### Quantifiers

```python
# Format -> e{m,n}
print(re.findall("A{2,5}",grades))

# if provided one value in braces, taken as both values
print(re.findall("B{3}",grades))

# de-streak
print(re.findall("A{1,5}B{1,5}C{1,5}",grades))
```

### Regex for Data Cleaning

```python
# Regex - Regular Expression, used to clean textual data. We can write our own regex expression
import re

# To find pattern string at starting 
main = "This is a Good Day :)"
if re.match("This",main):
    print("Yes")

# To find anywhere in the main string
if re.search("Good",main):
    print("Yes")
```

```python
# Just like checking using Conditionals, we can Segment a String.
# Tokenizing -> Splitting a string into substrings based on patterns.
# Tokenizing is used heavily in NLP

# Using split()
main = "Shreyas is a Coder. Shreyas is also a Musician. Shreyas has a great life."
print(re.split('Shreyas',main))

# Using findall()
print(re.findall('Shreyas', main))
```

### Merging One or More DataFrames

Full Outer Join -> Union<br>
Inner Join -> Intersection

```python
import pandas as pd


stdf = pd.DataFrame([{'Name': 'Kelly', 'Role': 'Director of HR'},
                         {'Name': 'Sally', 'Role': 'Course liasion'},
                         {'Name': 'James', 'Role': 'Grader'}])

stdf = stdf.set_index('Name')

sdf = pd.DataFrame([{'Name': 'James', 'School': 'Business'},
                           {'Name': 'Mike', 'School': 'Law'},
                           {'Name': 'Sally', 'School': 'Engineering'}])

sdf = sdf.set_index('Name')
```

```python
stdf # Staff
```

```python
sdf # Student
```

```python
# James and Sally are both Students and Staff, but Mike is not

# Creating Union
pd.merge(stdf,sdf, how='outer', left_index=True, right_index=True)
```

```python
# Intersection
pd.merge(stdf,sdf, how='inner', left_index=True, right_index=True)
```

```python
# left-Join
pd.merge(stdf,sdf, how='left', left_index=True, right_index=True)
```

```python
# Right-Join
pd.merge(stdf,sdf, how='right', left_index=True, right_index=True)
```

```python
# We need not use Indexes to merge df, param called "on"

# Remove index
stdf = stdf.reset_index()
sdf = sdf.reset_index()

# Join based on name column
pd.merge(stdf, sdf, how='right', on='Name') # Mistake
```

```python
stdf = stdf.reset_index()
stdf
```

```python
sdf = sdf.reset_index()
sdf
```

```python
pd.merge(stdf, sdf, how='right', on='Name')

# Always reset the index of dataframe before merging them
```

### Handling Conflicts between df

```python
stdf = pd.DataFrame(
    [
        {"Name": "Kelly", "Role": "Director of HR", "Location": "State Street"},
        {"Name": "Sally", "Role": "Course liasion", "Location": "Washington Avenue"},
        {"Name": "James", "Role": "Grader", "Location": "Washington Avenue"},
    ]
)
sdf = pd.DataFrame(
    [
        {"Name": "James", "School": "Business", "Location": "1024 Billiard Avenue"},
        {"Name": "Mike", "School": "Law", "Location": "Fraternity House #22"},
        {"Name": "Sally", "School": "Engineering", "Location": "512 Wilson Crescent"},
    ]
)
```

```python
stdf
```

```python
sdf
```

```python
# _x -> Left df index
# _y -> Right df index

# If we want all staff info regardless of whether they were students or not. But if they were students, we would want to get their student details as well -> Solution is left join

pd.merge(stdf, sdf, how='left', on='Name')
```

```python
# Multi-indexing and Multi-columns

stdf = pd.DataFrame(
    [
        {"First Name": "Kelly", "Last Name": "Desjardins", "Role": "Director of HR"},
        {"First Name": "Sally", "Last Name": "Brooks", "Role": "Course liasion"},
        {"First Name": "James", "Last Name": "Wilde", "Role": "Grader"},
    ]
)
sdf = pd.DataFrame(
    [
        {"First Name": "James", "Last Name": "Hammond", "School": "Business"},
        {"First Name": "Mike", "Last Name": "Smith", "School": "Law"},
        {"First Name": "Sally", "Last Name": "Brooks", "School": "Engineering"},
    ]
)
```

```python
stdf
```

```python
sdf
```

```python
# James Wilde and James Hammond are separate names hence won't be included in Inner Join

pd.merge(stdf, sdf, how='inner', on=['First Name', 'Last Name']) # Use list of columns
```

### Concatenating DataFrames <br>
Joining -> Horizontally <br>
Concatenating -> Vertically 

```python
%%capture

df_2011 = pd.read_csv(
    "datasets/college_scorecard/MERGED2011_12_PP.csv",
    on_bad_lines='skip'
)

df_2012 = pd.read_csv(
    "datasets/college_scorecard/MERGED2012_13_PP.csv",
    on_bad_lines='skip'
)

df_2013 = pd.read_csv(
    "datasets/college_scorecard/MERGED2013_14_PP.csv",
    on_bad_lines='skip'
)
```

```python
df_2011.head(3)
```

```python
print(len(df_2011))
print(len(df_2012))
print(len(df_2013))
```

```python
frames = [df_2011, df_2012, df_2013]
len(pd.concat(frames))
```

```python
# Concatenation is Successful
# Problem is that we don't know which data is from which layer
# Hence we can set an extra level of index

pd.concat(frames, keys=['2011', '2012', '2013']).sample(10)
```

```python
# Concatenation operation also have inner and outer methods
# If all df do not have identical columns -> Outer Join
# If all df needs to have only identical columns -> Inner Join
```

### 1. Method Chaining 

Every method on an object returns a reference to that object. Hence you can condense many different operations on a DataFrame.

```python
# Pull out the state and city names as a multiple index, and only for data which has a summary level of 50, which in this dataset is county-level data. Rename the column too to make it more readable.

(
    df
        .loc[
            df['SUMLEV'] == 50,
            ['STNAME', 'CTYNAME', 'ESTIMATESBASE2010']
        ]
        .dropna()
        .set_index(['STNAME', 'CTYNAME'])
        .rename(columns={'ESTIMATESBASE2010' : 'Estimates Base 2010'})
        .sample(15)
)
```

```python
def min_max(row):
    data = row[['POPESTIMATE2010',
                'POPESTIMATE2011',
                'POPESTIMATE2012',
                'POPESTIMATE2013',
                'POPESTIMATE2014',
                'POPESTIMATE2015']]
    
    row['max'] = np.max(data)
    row['min'] = np.min(data)
    return row

df.apply(min_max, axis='columns').sample(10)
```

```python
def get_state_region(x):
    northeast = [
        "Connecticut",
        "Maine",
        "Massachusetts",
        "New Hampshire",
        "Rhode Island",
        "Vermont",
        "New York",
        "New Jersey",
        "Pennsylvania",
    ]
    midwest = [
        "Illinois",
        "Indiana",
        "Michigan",
        "Ohio",
        "Wisconsin",
        "Iowa",
        "Kansas",
        "Minnesota",
        "Missouri",
        "Nebraska",
        "North Dakota",
        "South Dakota",
    ]
    south = [
        "Delaware",
        "Florida",
        "Georgia",
        "Maryland",
        "North Carolina",
        "South Carolina",
        "Virginia",
        "District of Columbia",
        "West Virginia",
        "Alabama",
        "Kentucky",
        "Mississippi",
        "Tennessee",
        "Arkansas",
        "Louisiana",
        "Oklahoma",
        "Texas",
    ]
    west = [
        "Arizona",
        "Colorado",
        "Idaho",
        "Montana",
        "Nevada",
        "New Mexico",
        "Utah",
        "Wyoming",
        "Alaska",
        "California",
        "Hawaii",
        "Oregon",
        "Washington",
    ]

    if x in northeast:
        return "Northeast"
    elif x in midwest:
        return "Midwest"
    elif x in south:
        return "South"
    else:
        return "West"
```

```python
df['region'] = df['STNAME'].apply(lambda x : get_state_region(x))

df[['STNAME', 'region']].sample(10)
```

```python
# Chained Version
(
    df
    .assign(
        region = df['STNAME'].apply(get_state_region)
    )

    [['STNAME', 'region']]

    .sample(10)
)
```

```python
# Self - Learning Groupby Iteration Method
data = pd.read_csv('data.csv')
data = data.set_index('Index')
data.columns = [x.lower().strip() for x in data.columns]
data.head()
```

```python
data['subscription date'] = pd.to_datetime(data['subscription date']).dt.year
data['subscription date'].head()
```

```python
for val,frame in data.sample(10).groupby('subscription date'):
    print('YEAR : ', val)
    print('----------------------------------------------------------------')
    print('DATAFRAME : \n', frame)
    print('----------------------------------------------------------------')
    print('----------------------------------------------------------------')
    print('\n')
```

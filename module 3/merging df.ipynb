### Merging One or More DataFrames

Full Outer Join -> Union<br>
Inner Join -> Intersection

```python
import pandas as pd


stdf = pd.DataFrame([{'Name': 'Kelly', 'Role': 'Director of HR'},
                         {'Name': 'Sally', 'Role': 'Course liasion'},
                         {'Name': 'James', 'Role': 'Grader'}])

stdf = stdf.set_index('Name')

sdf = pd.DataFrame([{'Name': 'James', 'School': 'Business'},
                           {'Name': 'Mike', 'School': 'Law'},
                           {'Name': 'Sally', 'School': 'Engineering'}])

sdf = sdf.set_index('Name')
```

```python
stdf # Staff
```

```python
sdf # Student
```

```python
# James and Sally are both Students and Staff, but Mike is not

# Creating Union
pd.merge(stdf,sdf, how='outer', left_index=True, right_index=True)
```

```python
# Intersection
pd.merge(stdf,sdf, how='inner', left_index=True, right_index=True)
```

```python
# left-Join
pd.merge(stdf,sdf, how='left', left_index=True, right_index=True)
```

```python
# Right-Join
pd.merge(stdf,sdf, how='right', left_index=True, right_index=True)
```

## 1) Ratio Scale :
- Units are Equally Spaced <br>
- All mathematical operations such as +, -, *, / are valid <br>
- Contains abscence of a value like 0 <br>
- Ex : Height, Weight <br>

## 2) Interval Scale :
- Units are Equally Spaced <br>
- No abscence of value, No absolute 0 <br>
- Some mathematical operations are invalid like division <br>
- Ex : Temperature, Direction <br>

## 3) Ordinal Scale :
- Order of Units is important <br>
- Not evenly spaced <br>
- Ex : Grades such as A+, A, B etc <br>

## 4) Nominal Scale :
- Just categories of data, no order <br>
- Most observed <br>
- Ex : Teams of a sports team <br>

```python
import pandas as pd
df=pd.DataFrame(['A+', 'A', 'A-', 'B+', 'B', 'B-', 'C+', 'C', 'C-', 'D+', 'D'],
                index=['excellent', 'excellent', 'excellent', 'good', 'good', 'good', 
                       'ok', 'ok', 'ok', 'poor', 'poor'],
               columns=["Grades"])
df
```

```python
df.dtypes
```

```python
df['Grades'].astype("category")
```

```python
my_categories = pd.CategoricalDtype(categories=['D', 'D+', 'C-', 'C', 'C+', 'B-', 'B', 'B+', 'A-', 'A', 'A+'], 
                           ordered=True)

# then we can just pass this to the astype() function
grades=df["Grades"].astype(my_categories)
grades.head()
```

```python
# Advantages -> Order for boolean masking
df[df['Grades'] > 'C']
```

```python
grades[grades > 'C']
```

```python
# Variables with boolean values are typically called as dummy variables
# Built-in function -> get_dummies
# Cut function to make data categorical

grades = grades.to_frame('Values')
grades
```

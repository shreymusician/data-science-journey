```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('datasets/cwurData.csv')
df.head()
```

```python
def assign(rank):
    if rank in range(1,101):
        return "First Tier Top University"

    elif rank in range(101, 201):
        return "Second Tier Top University"

    elif rank in range(201, 301):
        return "Third Tier Top University"

    else:
        return "Other Top University"

df['Rank_Level'] = df['world_rank'].apply(assign)

df.sample(10)
```

## Pivot Table

```python
# Compare rank level versus country of the universities in terms of overall score

# Values -> Score, Index -> Country, Columns -> Rank levels

df.pivot_table(values='score', index='country', columns='Rank_Level', aggfunc='mean').head()
```

```python
df.pivot_table(values='score', index='country', columns='Rank_Level', aggfunc=['mean', 'max']).head()
```

```python
df.pivot_table(values='score', index='country', columns='Rank_Level', aggfunc=['mean', 'max'], margins=True).head()
```

```python
ndf = df.pivot_table(values='score', index='country', columns='Rank_Level', aggfunc=['mean', 'max'], margins=True).head()

ndf.index
```
